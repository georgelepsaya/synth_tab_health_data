```python
from azure.ai.ml import command, Input, Output

command_job = command(
    code="./jobs",
    command="python run_experiment.py ${{inputs.model_name}} ${{inputs.dataset_path}} ${{outputs.metrics_report}}",
    environment="synthtabhealthdata:4",
    inputs={
        "model_name": "rtvae",
        "dataset_path": Input(
            type="uri_file",
            path="azureml:heart_disease:1",
        ),
    },
    outputs={
        "metrics_report": Output(type="uri_file", mode="rw_mount"),
    },
    compute="cpu-cluster",
)
```

```python
from azure_ml_client import get_ml_client

ml_client = get_ml_client()

# submit the command
returned_job = ml_client.jobs.create_or_update(command_job)

# get a URL for the status of the job
returned_job.studio_url
```
