Prepare Azure ML client:


```python
from azureml.azure_ml_client import get_ml_client

ml_client = get_ml_client()
```

```python
from azure.ai.ml import load_job
```

### ADS-GAN evaluation job


```python
dataset_names = [
    'alzheimers',
    'breast_cancer',
    'coronary_artery',
    'diabetes',
    'gallstone'
]
```

```python
for name in dataset_names:
    job_def = load_job(source=f"./jobs/adsgan/adsgan_{name}.yml")
    job = ml_client.jobs.create_or_update(job_def)
    print(f"Job {job.name} submitted: {job.studio_url}")
```

```python
# resubmit
job_def = load_job(source="./jobs/adsgan/adsgan_breast_cancer.yml")
job = ml_client.jobs.create_or_update(job_def)
print(f"Job {job.name} submitted: {job.studio_url}")
```

```python
job_cardio_train_adsgan_def = load_job(source="./jobs/adsgan/adsgan_cardio_train.yml")
job_cardio_train_adsgan = ml_client.jobs.create_or_update(job_cardio_train_adsgan_def)
print(f"Job {job_cardio_train_adsgan.name} submitted: {job_cardio_train_adsgan.studio_url}")
```
