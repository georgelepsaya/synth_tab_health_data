Prepare Azure ML client:


```python
from azureml.azure_ml_client import get_ml_client

ml_client = get_ml_client()
```

```python
from azure.ai.ml import load_job
```

### ADS-GAN evaluation job


```python
job_def = load_job(source="./jobs/evaluate_adsgan.yml")
```

```python
print(job_def)
```

```python
job = ml_client.jobs.create_or_update(job_def)
print(f"Job {job.name} submitted: {job.studio_url}")
```
